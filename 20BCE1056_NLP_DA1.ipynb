Parser

```python
 from nltk import CFG
 grammar = CFG.fromstring(""" S -> NP VP
  PP -> P NP
 NP -> Det N | NP PP
 VP -> V NP | VP PP
Det -> 'a' | 'the'
 N -> 'dog' | 'cat'
 V -> 'chased' | 'sat'
 P -> 'on' 
 """)
```

```python
grammar.start()
```

```python
grammar.productions()
```

```python
data = ['a','dog','chased','the','cat']
p = nltk.ChartParser(grammar)
for t in p.parse(data):
  print(t)
```

```python
gr = nltk.CFG.fromstring(""" S -> NP VP
 PP -> P NP
 NP -> Det N | Det N PP | 'I'
 VP -> V NP | VP PP
 Det -> 'an' 
 N -> 'elephant'
 V -> 'shot'
 P -> 'in'
 """)
sent = ['I', 'shot', 'an', 'elephant']
parser = nltk.ChartParser(gr)
for tree in parser.parse(sent):
  print(tree)
```

Word segmentation

```python
txt="Travelling is my hobby. Going to different places and exploring them is something that I always love to. More so, when you travel to new places, you meet new people, get to know about their culture, their food habits, dressing styles and much more. Trying delicious legacies of different places is amazing. One also gets to know about the historical importance of places, monuments situated and what not! Not only all this, I even love visiting beaches. During sunset, when the bright sun rays fall over the sea/ocean, it looks so beautiful and admirable. Even the zoos, which are places where most of the animals are found, are amongst one of the places I love to go. Walking around, looking at the lovely animals playing with each other, and feeding the birds and the cute puppies is a pleasure in its own. Even the monuments at different places tell about the history of that particular place like why it was built, when, who built and the entire story behind. More than reading about it, I love to visit such places and know the story behind. God has created such a beautiful world; it wouldn't be fair not to visit such amazing places. Hence, one should always travel and spend time with family as well as friends whenever possible."
print(word_tokenize(txt))
print(len(word_tokenize(txt)))
```

Sentence segmentation

```python
txt="Travelling is my hobby. Going to different places and exploring them is something that I always love to. More so, when you travel to new places, you meet new people, get to know about their culture, their food habits, dressing styles and much more. Trying delicious legacies of different places is amazing. One also gets to know about the historical importance of places, monuments situated and what not! Not only all this, I even love visiting beaches. During sunset, when the bright sun rays fall over the sea/ocean, it looks so beautiful and admirable. Even the zoos, which are places where most of the animals are found, are amongst one of the places I love to go. Walking around, looking at the lovely animals playing with each other, and feeding the birds and the cute puppies is a pleasure in its own. Even the monuments at different places tell about the history of that particular place like why it was built, when, who built and the entire story behind. More than reading about it, I love to visit such places and know the story behind. God has created such a beautiful world; it wouldn't be fair not to visit such amazing places. Hence, one should always travel and spend time with family as well as friends whenever possible."
print(sent_tokenize(txt))
print(len(sent_tokenize(txt)))
```

Convert to Lowercase

```python
lower_text = txt.lower()
print(lower_text)
```

Stop words removal

```python
# nltk.download('stopwords')
from nltk.corpus import stopwords
english_stopwords = stopwords.words('english')
tokens = word_tokenize(txt.lower())
tokens_without_stopwords = [t for t in tokens if t not in english_stopwords]
print(tokens_without_stopwords)
print(" ".join(tokens_without_stopwords))
```

Stemming

```python
from nltk.stem import PorterStemmer
e_words= word_tokenize(txt)
ps =PorterStemmer()
for w in e_words:
    rootWord=ps.stem(w)
    print(rootWord)
```

Lemmatization

```python
# nltk.download('wordnet')
# nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()
tokenization = nltk.word_tokenize(txt)
for w in tokenization:
  root=lem.lemmatize(w)
  print(root)
```

Part of speech tagger

```python
nltk.pos_tag(word_tokenize(txt))
```
